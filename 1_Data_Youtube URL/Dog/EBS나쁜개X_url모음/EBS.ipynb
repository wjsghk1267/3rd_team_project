```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd

# ChromeDriver 초기화
driver = webdriver.Chrome()

# 해당 페이지로 이동
driver.get('https://www.youtube.com/playlist?list=PLpuzWnAKjQgBBH113_kOsBGCZO8MVNazu')

# 페이지가 로드될 시간을 기다림
time.sleep(3)

# 페이지 끝까지 스크롤을 내리는 함수
def scroll_to_bottom(driver):
    last_height = driver.execute_script("return document.documentElement.scrollHeight")
    while True:
        # 페이지의 맨 아래까지 스크롤
        driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
        
        # 새로운 콘텐츠가 로드될 시간을 기다림
        time.sleep(2)

        # 스크롤 후 새로운 높이를 확인
        new_height = driver.execute_script("return document.documentElement.scrollHeight")
        
        # 더 이상 새로운 콘텐츠가 없으면 스크롤을 멈춤
        if new_height == last_height:
            break
        
        last_height = new_height

# 페이지 끝까지 스크롤
scroll_to_bottom(driver)

# 모든 video-title 요소를 가져옴
video_titles = driver.find_elements(By.XPATH, '//*[@id="video-title"]')

# 데이터를 저장할 리스트
data = []

# 각 영상 제목과 href를 추출하여 리스트에 추가
for video in video_titles:
    title = video.get_attribute('title')  # 제목을 가져옴
    href = video.get_attribute('href')    # href 링크를 가져옴
    data.append({'제목': title, '링크': href})

# 데이터프레임 생성
df = pd.DataFrame(data)

# CSV 파일로 저장
df.to_csv("반려견의 모든것 ! '세상에 나쁜 개는 없다'.csv", index=False, encoding='utf-8-sig')

# 드라이버 종료
driver.quit()
```
